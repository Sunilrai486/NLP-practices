```python
from nltk.corpus import webtext
```

```python
words = [w.lower() for w in webtext.words('grail.txt')]
```

```python
output = ' '.join(words)
```

```python
print(output)
```

```python
len(output)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation

listdatas = word_tokenize(output)
stopwords = stopwords.words('english') + list(punctuation)
filteredlist = []
for listdata in listdatas:
    if(listdata not in stopwords):
        filteredlist.append(listdata)
filteredlist
```

```python
len(filteredlist)
```

```python
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

words = [w.lower() for w in webtext.words('grail.txt')]
bcf = BigramCollocationFinder.from_words(words)

data = bcf.nbest(BigramAssocMeasures.likelihood_ratio, 10)

print(data)
```

```python
for dat in data:
    print(dat)
```

```python
from nltk.corpus import stopwords
stopset = set(stopwords.words('english'))
filter_stops = lambda w: len(w) < 3 or w in stopset
bcf = BigramCollocationFinder.from_words(words)

bcf.apply_word_filter(filter_stops)

n = 10

bcf.apply_freq_filter(n)

bcf.nbest(BigramAssocMeasures.likelihood_ratio, 4)
```

```python
from nltk.collocations import TrigramCollocationFinder
from nltk.metrics import TrigramAssocMeasures

words = [w.lower() for w in webtext.words('grail.txt')]
tcf = TrigramCollocationFinder.from_words(words)
tcf.apply_word_filter(filter_stops)

n = 5
tcf.apply_freq_filter(n)
tcf.nbest(TrigramAssocMeasures.likelihood_ratio, 10)
```

```python
import re
result = re.findall('clop.*?clop.*?clop', str(words))
len(result)
```

```python
import re
result = re.findall('clop clop clop', output)
len(result)
```

```python
from nltk.collocations import *
from nltk.tokenize import word_tokenize
text = "this is a foo bar bar black sheep  foo bar bar black sheep foo bar bar black sheep shep bar bar black sentence"
trigram_measures = nltk.collocations. TrigramAssocMeasures()
finder = TrigramCollocationFinder.from_words(word_tokenize(text))
for i in finder.score_ngrams(trigram_measures.pmi):
    print(i)
```

```python
from nltk.corpus import wordnet
```

```python
synonyms = []
antonyms = []
input_word = input()
for syn in wordnet.synsets(input_word):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print(set(synonyms))
print(set(antonyms))
```
